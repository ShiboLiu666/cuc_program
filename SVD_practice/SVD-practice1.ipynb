# 利用python实现SVD

```python
# coding=utf-8

#导入numpy库
from numpy import *
from numpy import linalg as la
```

```python
#导入原始数据
def loadExData():
     return [[1,1,1,0,0],
             [2,2,2,0,0],
             [5,5,5,0,0],
             [1,1,0,2,2],
             [0,0,0,3,3],
             [0,0,0,1,1]]
```

```python
Data = loadExData()
U,Sigma,VT = la.svd(Data)

print(U)
print(Sigma) #此时Sigma为列表
print(VT)
```

```python
#将列表Sigma转为矩阵Sig3
Sig3 = mat([[Sigma[0],0,0],[0,Sigma[1],0],[0,0,Sigma[2]]])
print (Sig3)

#重构原始矩阵
print (U[:,:3]*Sig3*VT[:3,:])
```
